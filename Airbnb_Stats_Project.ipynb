##**Statistical Analysis**

Is there a significant difference in the average room prices between different neighborhood groups?

Is there a significant difference in the average room prices between Airbnb rentals in Brooklyn and Manhattan?

Is there a significant association between the neighborhood group and the room type chosen by Airbnb guests?

```python
# # Steps of Hypothesis Testing

# #collect the data or sample
# Generate hypothesis:   Null Hypothesis :
#                        Alternate Hypothesis:
# #Choosing an appropriate test
# #test will return evidence in the form of p value
# if p value >0.05 :
#    accept the null hypothesis and reject the alternative
# else:
#    accept the alternative hypothesis and reject the null hypothesis
```

#Is there a significant difference in the average room prices between different neighborhood groups?


```python
#applying the ANNOVA test as the multiple samples are there
```

```python
# #collect the data or sample
df1=Airbnb_df[["price","neighbourhood_group"]]
df1
```

```python
groups=Airbnb_df["neighbourhood_group"].unique()
groups
```

```python
s1=df1[df1["neighbourhood_group"]=="Brooklyn"]["price"]
s1
```

```python
s2=df1[df1["neighbourhood_group"]=="Manhattan"]["price"]
s2
```

```python
s3=df1[df1["neighbourhood_group"]=="Queens"]["price"]
s3
```

```python
s4=df1[df1["neighbourhood_group"]=="Staten Island"]["price"]
s4
```

```python
s5=df1[df1["neighbourhood_group"]=="Bronx"]["price"]
s5
```

```python
final_data=[s1,s2,s3,s4,s5]
final_data
```

```python
# generating the hypothesis
#Null Hypothesis :  average price in all the groups is same
#alternate Hypothesis: average price in all the groups is not same
```

```python
# choosing test : we will choose annova as we are dealing with multiple samples
```

```python
from scipy.stats import f_oneway
```

```python
# Perform one-way ANOVA
statistic, p_value = f_oneway(*final_data)

# Display results
print("One-way ANOVA results:")
print(f"Statistic: {statistic}")
print(f"P-value: {p_value}")

```

```python
if p_value>0.05:
  print("accept the null hypothesis and reject the alternative")
else:
  print("accept the alternative hypothesis and reject the null hypothesis")
```

# Is there a significant difference in the average room prices between Airbnb rentals in Brooklyn and Manhattan?

```python
brooklyn_prices = df1[df1['neighbourhood_group'] == 'Brooklyn']['price']
manhattan_prices = df1[df1['neighbourhood_group'] == 'Manhattan']['price']
```

```python
brooklyn_prices
```

```python
manhattan_prices
```

```python
# null= avg prices of Both groups are same
# Alternative= avg prices of Both groups are  not same
```

```python
# z test -- if standard deviation is given
# t test      ---- if standard deviation is not given


# we will choose t test for 2 sample as std is not given
```

```python
from scipy.stats import ttest_ind
```

```python
brooklyn_prices.var()
```

```python
manhattan_prices.var()
```

```python
t_stats, p_value=ttest_ind(brooklyn_prices,manhattan_prices, equal_var=False)
```

```python
p_value
```

```python
if p_value>0.05:
  print("accept the null hypothesis and reject the alternative")
else:
  print("accept the alternative hypothesis and reject the null hypothesis")
```

# Is there a significant association between the neighborhood group and the room type chosen by Airbnb guests?

```python
# Null Hypothesis= the both the variables are independent having no association
# Alternate Hypothesis= the both the variables are dependent having association between them
```

```python
from scipy.stats import chi2_contingency
```

```python
contingency_table = pd.crosstab(Airbnb_df['neighbourhood_group'], Airbnb_df['room_type'])
```

```python
contingency_table
```

```python
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)
```

```python
p_value
```

```python
if p_value>0.05:
  print("accept the null hypothesis and reject the alternative")
else:
  print("accept the alternative hypothesis and reject the null hypothesis")
```
